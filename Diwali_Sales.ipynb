```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df.dropna(inplace = True)
```

```python
df['Amount'] = df['Amount'].astype(int)
```

Exploratory **Analysis**

```python
ax = sn.countplot(x = df['Gender'] , data = df)

for bar in ax.containers:
  ax.bar_label(bar)
```

```python
data = df.groupby('Gender', as_index = False).sum('Amount').sort_values(by= 'Amount', ascending = False)
sn.barplot(x= 'Gender' , y = 'Amount' , data = data)
```
